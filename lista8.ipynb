# Lista 8 - Porównanie z pracą konkursową

Wybrana lista: Lista 5 - metody zespołowe, klasyfikacja wieloklasowa (przewidywanie poziomu otyłości, zbiór UCI „Obesity or CVD risk", 7 klas).
Wybrany konkurs: Kaggle Playground Series S4E2 - Multi-Class Prediction of Obesity Risk

(https://www.kaggle.com/competitions/playground-series-s4e2)
Uzasadnienie wyboru:

To ten sam problem co w mojej liście - wieloklasowa klasyfikacja poziomu otyłości na 7 kategorii, na tych samych cechach (wiek, wzrost, waga, nawyki żywieniowe, aktywność). Co więcej, zbiór konkursowy został wygenerowany przez model uczony na dokładnie tym samym zbiorze UCI, którego użyłem w L5. Dzięki temu porównanie jest bezpośrednie - te same dane, ten sam cel, ta sama metryka sukcesu (accuracy).

## 1. Eksploracyjna analiza danych (EDA)

| Mój kod (L5) | Praca konkursowa (ddosad) |
|---|---|
| Sprawdziłem typy cech, braki danych, rozkład klas i podstawowe statystyki | To samo + bogate wizualizacje: histogramy każdej cechy, wykresy pudełkowe względem klasy, mapa korelacji |
| EDA była zwięzła, nastawiona na przygotowanie do modelu | EDA jest celem samym w sobie - pokazuje, że waga i wzrost niosą najwięcej informacji o klasie |
 
**Wniosek:** Mój EDA spełniał swoją rolę, ale był skromniejszy. Z pracy konkursowej przejąłbym systematyczne wizualizacje każdej cechy względem klasy docelowej - od razu widać, które cechy są najważniejsze, zanim w ogóle uruchomię model.

## 2. Preprocessing

| Mój kod (L5) | Praca konkursowa |
|---|---|
| Kodowanie kategorii przez `LabelEncoder` | Również kodują cechy kategoryczne i przygotowują dane pod modele gradient boosting |
 
**Wniosek:** Tu różnice są mniejsze niż w pozostałych obszarach. Warto zwrócić uwagę na sam dobór kodowania: `LabelEncoder` nadaje kategoriom sztuczną kolejność liczbową (0, 1, 2…). Dla modeli drzewiastych jest to akceptowalne, ale dla kategorii bez naturalnej kolejności one-hot bywa bezpieczniejszy. Dodatkowo rozważyłbym inżynierię cech - tworzenie nowych zmiennych łączących najważniejsze cechy (np. wagę i wzrost), co często ułatwia modelowi zadanie.

## 3. Optymalizacja hiperparametrów

| Mój kod (L5) | Praca konkursowa (ravi20076, lucamassaron) |
|---|---|
| Strojenie ręczne / `GridSearchCV` na kilku wartościach | **Optuna** - automatyczne przeszukiwanie przestrzeni parametrów, oceniane na kroswalidacji |
| Kilka kombinacji dobieranych „na oko" | Optuna sama testuje wiele kombinacji i uczy się, które są obiecujące |
 
**Wniosek:** To wyraźna przewaga konkursu. Zamiast ręcznie wpisywać kilka wartości, ravi20076 i lucamassaron używają biblioteki **Optuna**, która automatycznie szuka najlepszych hiperparametrów i ocenia je na kroswalidacji. Do L5 przeniósłbym właśnie Optunę - mniej ręcznej roboty i większa szansa na trafienie dobrych ustawień.

## 4. Uczenie modelu

| Mój kod (L5) | Praca konkursowa (hardikgarg03 / ddosad) |
|---|---|
| Metody zespołowe ze sklearn: bagging, boosting, voting, stacking | Nowoczesne biblioteki gradient boosting: XGBoost, LightGBM, Random Forest |
| Porównanie kilku zespołów między sobą | Jeden mocny model boostingowy, mocno dostrojony, czasem łączenie kilku modeli |
 
**Wniosek:** Robiłem to samo w duchu (łączenie modeli), ale na bibliotekach sklearn. Konkurs pokazuje, że XGBoost i LightGBM dają lepsze wyniki „z pudełka" i są standardem dla danych tabelarycznych. Do L5 dorzuciłbym XGBoost jako dodatkowy model w porównaniu - zobaczyłbym, o ile bije moje zespoły ze sklearn.

## 5. Ewaluacja

| Mój kod (L5) | Praca konkursowa (ravi20076) |
|---|---|
| Ocena na jednym podziale (`train_test_split`) | Kroswalidacja - wynik uśredniany po kilku podziałach |
| Metryki: accuracy, recall, F1, pewność predykcji (confidence) | Głównie accuracy (metryka konkursu), liczona na kroswalidacji |
 
**Wniosek:** Moja ewaluacja patrzyła na więcej miar (recall, F1, pewność predykcji), co daje pełniejszy obraz jakości klasyfikacji niż samo accuracy. Za to konkurs ocenia model na kroswalidacji zamiast jednego podziału - wynik jest pewniejszy i mniej zależny od losowego rozbicia danych. Najlepsze byłoby połączenie obu: moje miary + ich kroswalidacja.

## Podsumowanie

**Co najsilniej wpływa na wyniki w tym problemie:**
1. **Dobór hiperparametrów** - automatyczne strojenie (Optuna) zamiast ręcznego.
2. **Wybór modelu** - XGBoost / LightGBM biją klasyczne zespoły ze sklearn.
3. **Solidna walidacja** - kroswalidacja zamiast jednego podziału.
Co przeniósłbym do swojej Listy 5:
Optunę do strojenia hiperparametrów, XGBoost/LightGBM jako modele, kroswalidację zamiast pojedynczego `train_test_split`, oraz rozważenie inżynierii cech i doboru kodowania.
